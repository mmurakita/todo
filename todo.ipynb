```python
import tkinter as tk

class App:
    def __init__(self):
        self.root=tk.Tk()
        self.vsb = tk.Scrollbar(orient="vertical", command=self.OnVsb)
        self.lb1 = tk.Listbox(self.root, yscrollcommand=self.vsb.set)
        self.lb2 = tk.Listbox(self.root, yscrollcommand=self.vsb.set)
        self.vsb.pack(side="right",fill="y")
        self.lb1.pack(side="left",fill="x", expand=True)
        self.lb2.pack(side="left",fill="x", expand=True)
        self.lb1.bind("<MouseWheel>", self.OnMouseWheel)
        self.lb2.bind("<MouseWheel>", self.OnMouseWheel)
        for i in range(100):
            self.lb1.insert("end","item %s" % i)
            self.lb2.insert("end","item %s" % i)
        self.root.mainloop()

    def OnVsb(self, *args):
        self.lb1.yview(*args)
        self.lb2.yview(*args)

    def OnMouseWheel(self, event):
        self.lb1.yview("scroll", event.delta,"units")
        self.lb2.yview("scroll",event.delta,"units")
        # this prevents default bindings from firing, which
        # would end up scrolling the widget twice
        return "break"

app=App()
```

```python
import tkinter as tk
import tkinter.ttk as ttk
# from tkinter import *
# tk=Tk()

class Application(tk.Frame):
    def __init__(self,master):
        super().__init__(master)
        self.grid()

        master.geometry("500x500")
        master.title("雛形")
        

        self.widget()
 

    def widget(self):
        
        self.titleframe = tk.Frame(self.master)        
        self.label_title = tk.Label(self.titleframe, text="title", width=20)
        self.label_title.grid(row=1,column=0,sticky=tk.EW)
        self.entry_title = tk.Entry(self.titleframe,  width=50)
        self.entry_title.grid(row=1,column=1,sticky=tk.EW)
        self.titleframe.grid(row=1,sticky=tk.EW)
        
        self.bodyframe = tk.Frame(self.master)        
        self.label_body = tk.Label(self.bodyframe, text="body", width=20)
        self.label_body.grid(row=2,column=0,sticky=tk.EW)
        self.selected_body= tk.Text(self.bodyframe, height=10, width=50, wrap=tk.CHAR)
        self.selected_body.grid(row=2,column=1,sticky=tk.EW)
        self.bodyframe.grid(row=2,sticky=tk.EW)
        
        
        #ボタンを作る
        self.button = tk.Button(self.master, text="ボタン",width=50)
        self.master.bind("<Key-Return>",self.SetValue) 
        self.button.bind("<Button-1>",self.SetValue) 

        self.button.grid(row=6,sticky=tk.EW)
        
        #値の初期化
        self.reset()       
        

        self.listbox =[]    
    
    # ScrollBar with Listbox 13
#         tk.Label(self.master, text="13: Listbox \n& ScrollBar").grid(row=8, column=0, sticky="ew")
        scrollbar_frame = tk.Frame(self.master)
        scrollbar_frame.grid(row=8, column=0, sticky=tk.EW)
        self.listbox2 = tk.Listbox(scrollbar_frame)
        for i in self.listbox:
            self.listbox2.insert(tk.END, i)
        self.listbox2.pack(fill = 'x')
#         scroll_bar =tk.Scrollbar(scrollbar_frame, command=self.listbox2.yview)
#         scroll_bar.pack(side=tk.RIGHT, fill=tk.Y)
#         self.listbox2.config(yscrollcommand=scroll_bar.set)        

    def reset(self):
        

        self.paramframe = tk.Frame(self.master)        
        self.paramframe.grid(row=0,sticky=tk.EW)
        
        
        level_box =(5,4,3,2,1)
        imp_combo = ttk.Combobox(self.paramframe, state='readonly')
        urg_combo = ttk.Combobox(self.paramframe, state='readonly')
        # リストの値を設定
        imp_combo["values"] = level_box
        urg_combo["values"] = level_box
        # デフォルトの値を食費(index=0)に設定
        imp_combo.current(0)
        urg_combo.current(0)

        

     
        self.imp_label = tk.Label(self.paramframe,text = "重要度：")
        
        self.urg_label = tk.Label(self.paramframe,text = "緊急度：")
        
        # 配置
        self.imp_label.grid(row=0,column=0,sticky=tk.EW)
        imp_combo.grid(row=0,column=1, sticky=tk.EW)
        self.urg_label.grid(row=0,column=2, sticky=tk.EW)
        urg_combo.grid(row=0,column=3, sticky=tk.EW)
        
        
        
        
    # 入力値をラベルにセットする．
    def SetValue(self, event):
        #ラベルを追加
        if self.entry_title.get() !="":
            self.listbox2.insert(tk.END, "・"+self.entry_title.get())
                
                
        
def main():
    win = tk.Tk()
    app = Application(master=win)
    app.mainloop()


    

if __name__ == "__main__":
    main()
```
