```python
import tkinter as tk
import tkinter as ttk
# from tkinter import *
# tk=Tk()

class Application(tk.Frame):
    def __init__(self,master):
        super().__init__(master)
        self.pack()

        master.geometry("300x500")
        master.title("雛形")
        

        self.widget()
        self.reset()        

    def widget(self):
        
        
        self.label_title = tk.Label(self.master, text="title", width=20)
        ## ラベルを配置
        self.label_title.pack(fill = "x")
        self.entry_title = tk.Entry(self.master,  width=50)
        self.entry_title.pack()
        
        self.label_body = tk.Label(self.master, text="body", width=20)
        self.label_body.pack()
        
        self.selected_body= tk.Text(self.master, height=10, width=50, wrap=tk.CHAR)
        self.selected_body.pack()

        #ボタンを作る
        self.button = tk.Button(self.master, text="ボタン",width=50)
        self.master.bind("<Key-Return>",self.SetValue) 
        self.button.bind("<Button-1>",self.SetValue) 

        self.button.place(x=105, y=60)
        self.button.pack()
        
        
        # Canvas Widget を生成
        self.canvas = tk.Canvas(self.master)
        self.canvas.pack()

                #TODOリストの表示
        self.todo_frame = tk.Frame(self.canvas, width=300, height=500,bg='Blue')
        
        # Scrollbar を生成して配置
        self.bar = tk.Scrollbar(self.todo_frame, orient=VERTICAL)
        self.bar.pack(side=tk.RIGHT, fill=tk.Y)

        # Scrollbarを制御をCanvasに通知する処理を追加
        self.bar.config(command=self.canvas.yview)

        # Canvasのスクロール範囲を設定
        self.canvas.config(scrollregion=(0,0,100,100))

        # Canvasの可動域をScreoobarに通知する処理を追加
        self.canvas.config(yscrollcommand=self.bar.set)
        

        self.canvas.pack()

        

        self.todo_frame.pack(fill = "x",padx=10)
        self.todo_frame.pack_propagate(0)
        
        self.canvas.create_window((0,0), window=self.todo_frame, anchor=tk.NW, width=self.canvas.cget('width'))


    def reset(self):
        currencies = ('JPY', 'USD', 'EUR', 'CNY', 'MXN', 'CAD')
        v1 = StringVar(value=currencies)
        self.listbox = Listbox(self.todo_frame, height=3)  

        
    # 入力値をラベルにセットする．
    def SetValue(self, event):
        #ラベルを追加
        if self.entry_title.get() !="":
            self.label1 = tk.Label(self.todo_frame,text = self.entry_title.get())
            #表示
            self.label1.pack()

            self.listbox.insert(tk.END, self.entry_title.get())
            
            self.bar = ttk.Scrollbar(
                self.todo_frame, 
                orient=VERTICAL, 
                command=self.listbox.yview)
            self.listbox['yscrollcommand'] = self.bar.set
        
def main():
    win = tk.Tk()
    app = Application(master=win)
    app.mainloop()


    

if __name__ == "__main__":
    main()
```
